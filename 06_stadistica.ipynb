```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Simulacion de lanzamiento de un dado 100 veces

```python
sample = np.random.choice([1,2,3,4,5,6], 100)
```

Se necesita saber la cantidad de cada resultado

```python
from collections import Counter
c = Counter(sample)
```

```python
print({k: v/100.0 for k, v in c.items()})
```

Distribuciones continuas

[distribuciones](https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html)



Distribucion Normal
($\mu=0$ y  $\sigma^2=1$)

```python
norm_rv = sts.norm(0, 1)
sample = norm_rv.rvs(100)
```

Comparamos la distribucion empirica con la teorica

```python
x = np.linspace(-4,4,100)
cdf = norm_rv.cdf(x)
plt.plot(x, cdf, label='teorica')

from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label='empirica -ECDF')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='lower right')
```

```python
plt.hist(sample,bins=7, normed=True)
plt.ylabel('Frecuencia')
plt.xlabel('$x$')
```

Estimacion de la densidad empirica y comparacion la la densidad de la distribucion normal teorica

```python
df = pd.DataFrame(sample, columns=['KDE'])
ax = df.plot(kind='density')

x = np.linspace(-4,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='teorica pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
